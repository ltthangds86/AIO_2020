```
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```
# Data Preparation
(train_images, train_labels), (test_images, test_labels) = keras.datasets.fashion_mnist.load_data()

# Data Normalization [0,1]
train_images = train_images / 255.0
test_images  = test_images / 255.0 
```

```
# SGD + Sigmoid

model1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])

model1.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history_data1 = model1.fit(train_images, train_labels, 
                    validation_data=(test_images, test_labels), 
                    batch_size=1024, epochs=50, verbose=0)
```

```
# SGD + Tanh

# model
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='tanh'),
    keras.layers.Dense(10, activation='softmax')
])

model2.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history_data2 = model2.fit(train_images, train_labels, 
                    validation_data=(test_images, test_labels), 
                    batch_size=1024, epochs=50, verbose=0)
```

```
# Adam + Sigmoid

model3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])

model3.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history_data3 = model3.fit(train_images, train_labels, 
                    validation_data=(test_images, test_labels), 
                    batch_size=1024, epochs=50, verbose=0)
```

```
# Adam + Tanh

# model
model4 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='tanh'),
    keras.layers.Dense(10, activation='softmax')
])

model4.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history_data4 = model4.fit(train_images, train_labels, 
                    validation_data=(test_images, test_labels), 
                    batch_size=1024, epochs=50, verbose=0)
```

```
fig, axs = plt.subplots(2, 4, figsize=(20, 10))

axs[0, 0].plot(history_data1.history['loss'], label = "train_loss")
axs[0, 0].plot(history_data1.history['val_loss'], label = "val_loss")
axs[0, 0].set_ylabel('Loss')
axs[0, 0].set_title('SGD+Sigmoid')
axs[0, 0].legend()

axs[0, 1].plot(history_data2.history['loss'], label = "train_loss")
axs[0, 1].plot(history_data2.history['val_loss'], label = "val_loss")
axs[0, 1].set_title('SGD+Tanh')
axs[0, 1].legend()

axs[0, 2].plot(history_data3.history['loss'], label = "train_loss")
axs[0, 2].plot(history_data3.history['val_loss'], label = "val_loss")
axs[0, 2].set_title('Adam+Sigmoid')
axs[0, 2].legend()

axs[0, 3].plot(history_data4.history['loss'], label = "train_loss")
axs[0, 3].plot(history_data4.history['val_loss'], label = "val_loss")
axs[0, 3].set_title('Adam+Tanh')
axs[0, 3].legend()

axs[1, 0].plot(history_data1.history['accuracy'], label = "train_accuracy")
axs[1, 0].plot(history_data1.history['val_accuracy'], label = "val_accuracy")
axs[1, 0].set_xlabel('Iteration')
axs[1, 0].set_ylabel('Accuracy')
axs[1, 0].legend()

axs[1, 1].plot(history_data2.history['accuracy'], label = "train_accuracy")
axs[1, 1].plot(history_data2.history['val_accuracy'], label = "val_accuracy")
axs[1, 1].set_xlabel('Iteration')
axs[1, 1].legend()

axs[1, 2].plot(history_data3.history['accuracy'], label = "train_accuracy")
axs[1, 2].plot(history_data3.history['val_accuracy'], label = "val_accuracy")
axs[1, 2].set_xlabel('Iteration')
axs[1, 2].legend()

axs[1, 3].plot(history_data4.history['accuracy'], label = "train_accuracy")
axs[1, 3].plot(history_data4.history['val_accuracy'], label = "val_accuracy")
axs[1, 3].set_xlabel('Iteration')
axs[1, 3].legend()
```
