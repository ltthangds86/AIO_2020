```python
import matplotlib.pyplot as plt 
import numpy as np 

fig =plt.figure(figsize=(12,12))

for i in range(10):
    X_train_i = train_images[train_labels[:,0]==i]    
    X_train_i = X_train_i[:10]
    
    for j in range(10):
        img = X_train_i[j].reshape(32,32,3)
        fig.add_subplot(10, 10, i*10 + j+1)
        plt.axis('off')
        plt.imshow(img, vmin=0, vmax=255)
```

```python
# sigmoid and sgd

import tensorflow as tf
from tensorflow import keras

# Data Preparation
(train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()

# Data Normalization [0,1]
train_images = train_images / 255.0
test_images  = test_images / 255.0

# model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history_data = model.fit(train_images, train_labels, batch_size=1024, epochs=50)

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=0)
print('\n\nTest accuracy:', test_acc)
```
