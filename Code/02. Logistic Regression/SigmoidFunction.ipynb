```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_function(z):
    return 1 / (1 + np.exp(-z))

iris = np.genfromtxt('iris_1D.csv', dtype=None, delimiter=',', skip_header=1) 
X = iris[:, :1]
y = iris[:, 1]

print('X: \n', X)
print('y: \n', y)
```

```python
b_value  = -0.654 #-5.156
w_value  = 0.535  #2.331

z_values_t = w_value*X + b_value
y_values_t = sigmoid_function(z_values_t)

print('z_values: \n', z_values_t)
print('y_values: \n', y_values_t)
```

```python
x_values = np.arange(0,6,0.1)
z_values = w_value*x_values + b_value
y_values = sigmoid_function(z_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, z_values)
plt.plot(x_values, y_values, 'g')
plt.scatter(X[y == 0][:, 0], y[y == 0], color='b', label='0')
plt.scatter(X[y == 1][:, 0], y[y == 1], color='r', label='1')

plt.xlim(0, 6)
plt.ylim(-0.1, 1.1)
plt.legend()
plt.show()
```
